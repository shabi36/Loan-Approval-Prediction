```python
x = dataset.drop(columns="loan_status")
y = dataset["loan_status"]
```

```python
x
```

```python
y
```

```python
from imblearn.under_sampling import RandomUnderSampler  

ru = RandomUnderSampler()

ru_x , ru_y = ru.fit_resample(x,y)
```

```python
ru_x.shape
```

```python
ru_y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(ru_x , ru_y , test_size=0.2 , random_state=2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier,ExtraTreesClassifier
from sklearn.svm import SVC #support vector classifier
```

```python
from sklearn.metrics import  accuracy_score , confusion_matrix , precision_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import  StandardScaler
```

```python

knn= KNeighborsClassifier (n_neighbors=3)

pipe = Pipeline ([   
     ("step1" , StandardScaler()),
    ("step2" , knn)
])


pipe.fit (x_train , y_train)

y_predict = pipe.predict (x_test)


print("\nTRAIN ",pipe.score(x_train,y_train))
print("TEST", pipe.score(x_test,y_test))

print("\nCONFUSION MATRIX\n", confusion_matrix(y_test , y_predict))

print("\nACCURACY SCORE", accuracy_score(y_test , y_predict))
print("PRECISION SCORE", precision_score(y_test , y_predict , average=None))
```

```python
dt= DecisionTreeClassifier (max_depth= 4)

pipe = Pipeline ([
    
     ("step1" , StandardScaler()),
    ("step2" , dt)
])


pipe.fit (x_train , y_train)

y_predict = pipe.predict (x_test)



print("\nTRAIN ",pipe.score(x_train,y_train))
print("TEST", pipe.score(x_test,y_test))

print("\nCONFUSION MATRIX\n", confusion_matrix(y_test , y_predict))

print("\nACCURACY SCORE", accuracy_score(y_test , y_predict))
print("PRECISION SCORE", precision_score(y_test , y_predict ,average=None))
```

```python
rf= RandomForestClassifier (n_estimators=300,
                              random_state=3,
                              max_samples=0.5,
                              max_features=0.75,
                              max_depth=25)

pipe = Pipeline ([
   
     ("step1" , StandardScaler()),
    ("step2" , rf)
])


pipe.fit (x_train , y_train)

y_predict = pipe.predict (x_test)



print("\nTRAIN ",pipe.score(x_train,y_train))
print("TEST", pipe.score(x_test,y_test))

print("\nCONFUSION MATRIX\n", confusion_matrix(y_test , y_predict))

print("\nACCURACY SCORE", accuracy_score(y_test , y_predict))
print("PRECISION SCORE", precision_score(y_test , y_predict , average=None))
```
